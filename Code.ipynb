```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("heart1.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
one_hot_encode = pd.get_dummies(df,['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'])
```

```python
one_hot_encode.head()
```

```python
one_hot_encode.columns
```

```python
one_hot_encode=one_hot_encode[['Age', 'Sex_F', 'Sex_M', 'ChestPainType_ASY', 'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG_LVH', 'RestingECG_Normal', 'RestingECG_ST', 'MaxHR', 'ExerciseAngina_N', 'ExerciseAngina_Y', 'Oldpeak', 'ST_Slope_Down', 'ST_Slope_Flat', 'ST_Slope_Up' ,'HeartDisease']]
```

```python
one_hot_encode.columns
```

```python
y=one_hot_encode['HeartDisease']
```

```python
y.head()
```

```python
X=one_hot_encode.drop(["HeartDisease"], axis=1)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify= y, random_state=10)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.naive_bayes import GaussianNB          #applying Navie-bayes algorithm
gnb = GaussianNB()
```

```python
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix      #confusion matrix for y_test and y_pred
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
gnb.score(X_test,y_test)                           #having 87% accuracy for this data set
```
